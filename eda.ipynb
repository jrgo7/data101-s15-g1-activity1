```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from analyze import DATASET_URL, CATEGORY_COL, VALUE_COL
```

```python
df = pd.read_csv(DATASET_URL)
df.head()
```

```python
df.info()
```

```python
df_cleaned = df.dropna()
```

```python
df.describe()
```

```python
df.groupby(CATEGORY_COL).describe()
```

```python
sns.boxplot(x=CATEGORY_COL, y=VALUE_COL, data=df_cleaned, hue=CATEGORY_COL)
plt.title('Bill Length by Species')
plt.xlabel('Species')
plt.ylabel('Bill Length (mm)')
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df_cleaned, hue=CATEGORY_COL)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Bill Length vs Bill Depth by Species')
```
